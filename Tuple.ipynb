```python
#Creation of Tuple
citytuple=("Tiptur","Banglore","Dharwad","Tumkur","Gulbarga")
print(citytuple)
```

```python
#Accessing the Tuple
citytuple=("Tiptue","Banglore","Dharwad","Tumkur","Gulbarga")
print(citytuple)
print(citytuple[2])
print(citytuple[-2])
print(citytuple[2:5])
print(citytuple[-4:-1])
```

```python
#Change Tuple value
citytuple=("Tiptur","Banglore","Dharwad","Tumkur","Gulbarga")
print(citytuple)
c=list(citytuple)
c[1]="Davangere"
citytuple=tuple(c)
print(citytuple)
```

```python
#Use of Loop in Tuple
citytuple=("Tiptur","Banglore","Dharwad","Tumkur","Gulbarga")
for x in citytuple:
  print(x)
```

```python
#Check Items exits
citytuple=("Tiptur","Banglore","Dharwad","Tumkur","Gulbarga")
key=input("Enter the city name")
if key in citytuple:
  print("Yes,key is in the tuple")
else:
  print("Key is not found in the tuple")  
```

```python
#Length of Tuple
citytuple=("Tiptur","Banglore","Dharwad","Tumkur","Gulbarga")
print(len(citytuple))
```

```python
#Add Items
citytuple=("Tiptur","Banglore","Dharwad","Tumkur","Gulbarga")
c=list(citytuple)
c.append("Mysore")
citytuple=tuple(c)
print(citytuple) 
```

```python
#Create Tuple of single Item
citytuple=("Tiptur",)
print(type(citytuple))
```

```python
#Removing Items
citytuple=("Tiptur","Banglore","Dharwad","Tumkur","Gulbarga")
print(citytuple)
del citytuple
print(citytuple)
```

```python
#Joining Two Tuples
citytuple=("Tiptur","Banglore","Dharwad")
citytuple1=("Mysore","Davangere","Mandya")
city_new=citytuple1+citytuple
print(city_new)
```

```python
#Count method
citytuple=("Tiptur","Banglore","Dharwad","Tumkur","Gulbarga")
print(citytuple.count("Tiptur"))
```

```python
#Index method
citytuple=("Tiptur","Banglore","Dharwad","Tumkur","Gulbarga")
print(citytuple.index("Dharwad"))
```

```python
#First element is number and Second element is square of the number
f_range=int (input("Enter the first element"))
s_range=int (input("Enter the second element"))
a=[(x,x**2) for x in range(f_range,s_range+1)]
print(a)
```

```python
#Eventuple and Oddtuple
atuple=()
A=list(atuple)

n=int (input("Enter number of elements:"))
for i in range(int(n)):
  b=int (input("Enter the elements:"))
  A.append(b)
even=()
odd=()  
Even=list(even)
Odd=list(odd)
for j in A:
  if (j%2==0):
   Even.append(j)
   
  else:
    Odd.append(j)
    
  
    atuple=tuple(A)
    eventuple=tuple(Even)
    

    print("Even tuple:",eventuple)
    oddtuple=tuple(Odd)
    print("Odd tuple:",oddtuple)
```
